```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
x=[1,5,1.5,8,1,8,5,6.1,3]
y=[2,8,1.8,8,0.6,10,2.1,4.7,8.9]
plt.scatter(x,y)
plt.show()
```

```python
x=np.array([[1,2],[5,8],[1.5,1.8],[8,8],[1,0.6],[8,10],[5,2.1],[6.1,4.7],[3,8.9]])
```

```python
kmeans=KMeans(n_clusters=3)
kmeans.fit(x)
centroids=kmeans.cluster_centers_
labels=kmeans.labels_
print(centroids)
print(labels)
```

```python
colors=['g.','r.','b.']
for i in range(len(x)):
    plt.plot(x[i][0],x[i][1],colors[labels[i]],markersize=20)
    plt.scatter(centroids[:,0],centroids[:,1],marker='*',s=100)
plt.show()
```
